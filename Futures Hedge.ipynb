```python
import numpy as np
import matplotlib.pyplot as plt

# Data arrays
weighted_return = np.array([0.000390367,0.754244974
,-0.314526579
,-0.32482639
,-0.217819389
,0.36836
,1.240452529
,0.775718618
,-0.457878525
,1.135176246
,-0.733717795
,0.438606967
,-0.155624894
,-0.330061016
,1.073335763
,0.279517255
,-0.026502587
,-0.712750523
,0.214139818
,0.096932822
,0.213658235
,0.444755695])
returns_market = np.array([0.007582142
,0.016147789
,0.00068716
,-0.015326967
,0.001414821
,-0.018459373
,-0.014478144
,-0.001509741
,0.016476648
,-0.006980856
,0.017561981
,-0.011019423
,0.008918359
,0.012982182
,-0.016463188
,0.002984529
,0.005639802
,0.001646944
,-0.001573841
,0.014237259
,0.005715265
,0.014436548])

# Calculate the covariance 
covariance = np.cov(weighted_return, returns_market)[0][1]

# Calculate the variance of the market returns
variance = np.var(returns_market, ddof=1)

# Beta is covariance divided by variance
beta = covariance / variance

# Create a scatter plot of market returns (x-axis) vs stock returns (y-axis)
plt.scatter(returns_market,weighted_return , label='Data points')

# Calculate the line of best fit, which will have a slope of the beta
y = beta * returns_market

# Plot the line of best fit
plt.plot(returns_market, y, color='red', label=f'Line of best fit (Beta = {beta:.2f})')

plt.xlabel('S&P 500 Returns')
plt.ylabel('ETF Weighted Return')
plt.title('S&P 500 Returns vs ETF Weighted Return')
plt.legend()
plt.show()
```
